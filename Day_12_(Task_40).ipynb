```python
# 1. Create a 2D list of contacts: [Name, Phone, Email]
contacts = [
    ["Devi", "9876543210", "devi@example.com"],
    ["Sethu", "9123456780", "sethu@example.com"],
    ["Chinchu", "9988776655", "chinchu@example.com"]
]
```

```python
# 2. Allow the user to add a new contact
print("Add a New Contact")
new_name = input("Enter name: ")
new_phone = input("Enter phone number: ")
new_email = input("Enter email: ")
```

```python
# Append new contact as a list
contacts.append([new_name, new_phone, new_email])
```

```python
# 3. Display all contacts sorted alphabetically by name
# Sort using key → contact[0] (name)
contacts.sort(key=lambda x: x[0])

print("\nAll Contacts Sorted by Name:")
for c in contacts:
    print(c)
```

```python
# 4. Search for contacts by name (partial match)
search_term = input("\nEnter a name to search: ").lower()

print("\nSearch Results:")
found = False
for c in contacts:
    if search_term in c[0].lower():   # partial match
        print(c)
        found = True

if not found:
    print("No matching contacts found.")
```

```python
# 5. Display only the first 3 contacts after sorting
print("\nFirst 3 Contacts:")
for c in contacts[:3]:    # slicing
    print(c)
```
