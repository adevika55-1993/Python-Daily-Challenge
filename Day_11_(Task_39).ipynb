```python
# 1. Create a tuple with fixed product categories (immutable)
categories = ("Electronics", "Stationery", "Clothing")
```

```python
# 2. Create a list of products that can be updated (mutable)
products = ["Mobile", "Pen", "Shirt"]
```

```python
# 3. Ask the user to enter a new product and add it to the list
new_product = input("Enter a new product to add: ")
products.append(new_product)   # append() adds item to the list
```

```python
# 4. Ask the user to input the index of a product to remove
print("Current product list with indexes:")
for i in range(len(products)):
    print(i, "-", products[i])
remove_index = int(input("Enter the index of the product to remove: "))
```

```python
# Check if the given index is valid
if 0 <= remove_index < len(products):
    removed_item = products.pop(remove_index)  # pop removes item at index
    print(f"Removed product: {removed_item}")
else:
    print("Invalid index! No product removed.")
```

```python
# 5. Display final list of products and categories
print("\nFinal Product List:", products)
print("Product Categories (Tuple):", categories)
```
