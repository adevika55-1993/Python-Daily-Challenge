```python
# BMI Calculator Program

# 1. Ask user to enter weight and height
weight = float(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in meters: "))
```

```python
# 2. Ensure the weight and height entered are positive
if weight <= 0 or height <= 0:
    print("Error: Weight and height must be positive numbers!")
else:
    # 3. Calculate BMI using formula
    bmi = weight / (height * height)
    # 4. Classify BMI using conditional statements
    if bmi < 18.5:
        category = "Underweight"
    elif 18.5 <= bmi < 24.9:
        category = "Normal weight"
    elif 25 <= bmi < 29.9:
        category = "Overweight"
    else:
        category = "Obese"
    # 5. Display the result clearly
    print(f"\nYour BMI is: {bmi:.2f}")
    print("Health Category:", category)
```
