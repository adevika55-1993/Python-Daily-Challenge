```python
# 1. Create a dictionary with student names and their grades
student_grades = {
    "Arun": 85,
    "Divya": 90,
    "Kavin": 78
}
```

```python
# 2. Allow the user to add or update grades for at least 3 students
print("Update/Add Grades for Students")

for i in range(3):
    name = input(f"Enter student name {i+1}: ")
    grade = int(input(f"Enter grade for {name}: "))
    student_grades[name] = grade   # adds new or updates existing
```

```python
# Display updated dictionary
print("\nUpdated Student Grades:")
for name, grade in student_grades.items():
    print(name, "-", grade)
```

```python
# 3. Create a set of hobbies and allow the user to input at least 3 hobbies
hobbies = set()
print("\nEnter 3 hobbies:")

for i in range(3):
    hobby = input(f"Hobby {i+1}: ")
    hobbies.add(hobby)
```

```python
# 4. Convert the set into a frozenset
frozen_hobbies = frozenset(hobbies)
```

```python
# Display results
print("\nHobbies (Set):", hobbies)
print("Hobbies (FrozenSet):", frozen_hobbies)
```
